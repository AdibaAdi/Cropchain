## Shared Agent Definitions

The Cropchain project uses a set of AutoGen agents aligned with core Waterfall software engineering roles:

- Customer Proxy
- Project Manager
- Requirements Engineer
- System Engineer
- Software Engineer
- Test Engineer
- Documentation Engineer

For **Experiment 1**, the active workflow focuses on the first three roles:
- Customer Proxy
- Project Manager
- Requirements Engineer

The remaining agents are defined here to preserve a consistent project-wide architecture for later experiments.

````python
customer_proxy_agent = autogen.ConversableAgent(
    name="Customer_Proxy",
    system_message="""
You represent the customer and business stakeholder.
Your job is to clearly describe the business need, users, pain points, and desired features.
Stay focused on business goals, expected benefits, and real-world use cases.
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

project_manager_agent = autogen.ConversableAgent(
    name="Project_Manager",
    system_message="""
You are the Project Manager for a Waterfall software planning process.

Your responsibilities:
1. understand the customer request
2. break the project into exactly these 5 Waterfall phases only:
   - Requirements
   - Analysis/Design
   - Implementation
   - Testing
   - Documentation
3. route the correct work to the correct engineer
4. ask for structured outputs
5. make sure each phase includes:
   - deliverables
   - tasks and activities
   - review tasks
   - rework tasks
   - productivity assumptions
   - effort estimates

Important rules:
- Do NOT introduce extra top-level phases such as Deployment, Maintenance, Operations, or Support.
- Keep the plan aligned with a graduate software project management course.
- Make the output structured, clear, and presentation-ready.
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

requirement_engineer_agent = autogen.ConversableAgent(
    name="Requirement_Engineer",
    system_message="""
You are the Requirements Engineer for a Waterfall project.

Your job:
1. generate detailed use cases for the requested system
2. number each use case uniquely like UC-001, UC-002
3. number each requirement uniquely like REQ-001, REQ-002
4. include both functional and non-functional requirements
5. focus on the specific domain, which is a farm-to-table crop supply chain platform

Your output must include:
A. use case list
B. detailed requirements
C. requirement phase tasks and activities
D. review tasks
E. rework tasks
F. productivity estimate
G. effort estimate
H. assumptions, risks, and open questions

Use these assumptions unless the prompt says otherwise:
- productivity rate = 5 requirements completed per day

Important:
- make the system domain-specific to Cropchain
- include actors such as Farmer, Restaurant Buyer, Grocery Buyer, Logistics Coordinator, and Platform Admin
- include AI-related features like yield prediction and fair pricing recommendation
- include traceability and supply chain visibility requirements
- make the output presentation-ready
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

system_engineer_agent = autogen.ConversableAgent(
    name="System_Engineer",
    system_message="""
You are the System Engineer for a Waterfall project.

Your future job in this project:
- convert requirements into design artifacts
- create architecture, data design, interfaces, workflow diagrams, security design, and documentation-oriented design outputs
- include design tasks, review tasks, rework tasks, productivity estimates, and effort estimates
- use 5 pages completed per day unless instructed otherwise
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

software_engineer_agent = autogen.ConversableAgent(
    name="Software_Engineer",
    system_message="""
You are the Software Engineer for a Waterfall project.

Your future job in this project:
- convert design into implementation components
- estimate source lines of code
- break implementation into modules
- include implementation tasks, review tasks, rework tasks, productivity estimates, and effort estimates
- use 50 SLOC completed per day unless instructed otherwise
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

test_engineer_agent = autogen.ConversableAgent(
    name="Test_Engineer",
    system_message="""
You are the Test Engineer for a Waterfall project.

Your future job in this project:
- create detailed test plans and test cases
- map tests to requirements
- include test execution tasks, review tasks, rework tasks, productivity estimates, and effort estimates
- use 2 test cases completed per day unless instructed otherwise
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

documentation_engineer_agent = autogen.ConversableAgent(
    name="Documentation_Engineer",
    system_message="""
You are the Documentation Engineer for a Waterfall project.

Your future job in this project:
- create user documentation, training materials, API docs, and release notes
- include documentation tasks, review tasks, rework tasks, productivity estimates, and effort estimates
- use 3 pages completed per day unless instructed otherwise
""",
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

print("All shared agents created successfully.")
````

## Prompt Design for Experiment 1

The following project-manager prompt is designed to guide the Requirements Engineer toward a structured, domain-specific, and presentation-ready output.

The prompt explicitly requests:
- uniquely identified use cases
- functional and non-functional requirements
- requirement-phase tasks and activities
- review and rework tasks
- productivity and effort estimates
- assumptions, risks, and open questions

This helps ensure that Experiment 1 produces a strong baseline artifact for the rest of the Waterfall workflow.

````python
project_manager_to_requirement_engineer_prompt = f"""
I have received the following customer request for the software project "{PROJECT_NAME}".

Customer request:
{CUSTOMER_MESSAGE}

Based on this request, generate a detailed Requirements Engineering output for a Waterfall project.

Your output must include:

1. At least 12 detailed use cases
   - Give every use case a unique ID such as UC-001
   - Include actor, goal, and short description

2. Detailed requirements
   - Give every requirement a unique ID such as REQ-001
   - Include both functional and non-functional requirements
   - Make them specific to Cropchain and the Food Services domain

3. Requirement phase tasks and activities
   Include:
   - stakeholder analysis
   - requirement elicitation
   - requirement analysis
   - use case drafting
   - requirement validation
   - requirement review
   - requirement rework
   - final requirements baseline

4. Productivity and effort estimation
   Use:
   - productivity rate = 5 requirements completed per day
   Show:
   - total work
   - productivity
   - estimated effort in days

5. Assumptions, risks, and open questions

6. Final structured summary that is ready for a class presentation

Important:
- Include actors such as Farmer, Restaurant Buyer, Grocery Buyer, Logistics Coordinator, and Platform Admin
- Include Cropchain-specific capabilities such as yield prediction, fair pricing, traceability, freshness monitoring, and shipment visibility
- Make the output correct, complete, and coherent
""".strip()

print(project_manager_to_requirement_engineer_prompt[:900] + "\n...")
````

## Output Persistence Helpers

The following helper functions save generated outputs and chat histories into the Experiment 1 output folder.

This supports:
- reproducibility
- artifact reuse in later experiments
- final presentation preparation
- experiment comparison across project phases

````python
def save_text(filename: str, text: str):
    file_path = OUTPUT_DIR / filename
    file_path.write_text(text, encoding="utf-8")
    print(f"Saved: {file_path}")

def save_chat_history(filename: str, chat_result):
    file_path = OUTPUT_DIR / filename

    if hasattr(chat_result, "chat_history"):
        lines = []
        for item in chat_result.chat_history:
            role = item.get("name", item.get("role", "Unknown"))
            content = item.get("content", "")
            lines.append(f"{role}:\n{content}\n")

        file_path.write_text(("\n" + ("-" * 80) + "\n").join(lines), encoding="utf-8")
        print(f"Saved chat history: {file_path}")
    else:
        print("No chat_history found on chat result.")
````

## Experiment 1 Objective

Experiment 1 focuses on the early stages of Waterfall planning for Cropchain. Its purpose is to transform the initial business problem into a structured requirements baseline.

### Focus Areas
- Business problem interpretation
- Early scope clarification
- Requirements generation
- Requirement-phase task definition
- Initial effort estimation

### Why This Experiment Matters
The outputs generated here serve as the baseline planning artifacts for the downstream design, implementation, testing, and documentation phases.

````python
def run_experiment_1():
    print("=" * 80)
    print("STEP 1: Customer Proxy -> Project Manager")
    print("=" * 80)

    customer_to_pm = customer_proxy_agent.initiate_chat(
        recipient=project_manager_agent,
        message=CUSTOMER_MESSAGE,
        max_turns=1,
    )

    print("\n" + "=" * 80)
    print("STEP 2: Project Manager -> Requirement Engineer")
    print("=" * 80)

    pm_to_re = project_manager_agent.initiate_chat(
        recipient=requirement_engineer_agent,
        message=project_manager_to_requirement_engineer_prompt,
        max_turns=1,
    )

    return customer_to_pm, pm_to_re
````

````python
# Run this cell when you are ready to generate the outputs.
customer_to_pm_result, pm_to_re_result = run_experiment_1()
````

## Saving Experiment 1 Artifacts

To support later project phases, the final outputs from the Project Manager and Requirements Engineer are saved as reusable markdown artifacts. Raw chat histories are also preserved for traceability.

````python
# Save raw chat histories
save_chat_history("01_customer_to_pm_chat.txt", customer_to_pm_result)
save_chat_history("02_pm_to_re_chat.txt", pm_to_re_result)

# Save cleaner final artifacts for downstream project reuse
if hasattr(customer_to_pm_result, "chat_history") and customer_to_pm_result.chat_history:
    pm_output = customer_to_pm_result.chat_history[-1].get("content", "")
    save_text("03_project_manager_summary.md", pm_output)

if hasattr(pm_to_re_result, "chat_history") and pm_to_re_result.chat_history:
    re_output = pm_to_re_result.chat_history[-1].get("content", "")
    save_text("04_requirements_engineer_output.md", re_output)

print("Experiment 1 artifacts saved.")
````
