```python
import os
from pathlib import Path
from dotenv import load_dotenv
import autogen

CURRENT_DIR = Path.cwd()
BASE_DIR = CURRENT_DIR.parent if CURRENT_DIR.name == "notebooks" else CURRENT_DIR

load_dotenv(BASE_DIR / ".env", override=True)

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
if not OPENAI_API_KEY:
    raise ValueError("OPENAI_API_KEY is missing. Add it to your .env file before running.")

MODEL_NAME = "gpt-4o-mini"

TEAM_LLM_CONFIG = {
    "config_list": [
        {
            "model": MODEL_NAME,
            "api_key": OPENAI_API_KEY,
        }
    ],
    "temperature": 0.2,
    "timeout": 120,
}

PROJECT_NAME = "Cropchain: Intelligent Farm-to-Table Supply Chain Management System"

EXP1_DIR = BASE_DIR / "src" / "outputs" / "experiment_1"
EXP2_DIR = BASE_DIR / "src" / "outputs" / "experiment_2"
OUTPUT_DIR = BASE_DIR / "src" / "outputs" / "experiment_3"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

PM_SUMMARY_FILE = EXP1_DIR / "03_project_manager_summary.md"
REQ_OUTPUT_FILE = EXP1_DIR / "04_requirements_engineer_output.md"
DESIGN_OUTPUT_FILE = EXP2_DIR / "03_system_engineer_output.md"
IMPLEMENTATION_OUTPUT_FILE = EXP2_DIR / "04_software_engineer_output.md"

required_files = [
    PM_SUMMARY_FILE,
    REQ_OUTPUT_FILE,
    DESIGN_OUTPUT_FILE,
    IMPLEMENTATION_OUTPUT_FILE,
]

for f in required_files:
    if not f.exists():
        raise FileNotFoundError(f"Required upstream file is missing: {f}")

pm_summary_text = PM_SUMMARY_FILE.read_text(encoding="utf-8")
requirements_text = REQ_OUTPUT_FILE.read_text(encoding="utf-8")
design_text = DESIGN_OUTPUT_FILE.read_text(encoding="utf-8")
implementation_text = IMPLEMENTATION_OUTPUT_FILE.read_text(encoding="utf-8")

print("Experiment 3 environment ready.")
print("Model:", MODEL_NAME)
print("Loaded all upstream Waterfall artifacts successfully.")
```

```python
project_manager_agent = autogen.ConversableAgent(
    name="Project_Manager",
    system_message="""
You are the Project Manager for a Waterfall software planning process.
Keep the outputs structured, clear, and presentation-ready.
Always request review tasks, rework tasks, effort, assumptions, risks, and open questions.
""".strip(),
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

test_engineer_agent = autogen.ConversableAgent(
    name="Test_Engineer",
    system_message="""
You are the Test Engineer for Cropchain.
Create a detailed testing output that includes:
- mapped test cases
- coverage of critical requirements
- functional and non-functional testing areas
- alerts, pricing, prediction, logistics visibility, and traceability validation
- test execution tasks
- review tasks
- rework tasks
- productivity and effort estimates
Use 2 test cases completed per day unless instructed otherwise.
""".strip(),
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

documentation_engineer_agent = autogen.ConversableAgent(
    name="Documentation_Engineer",
    system_message="""
You are the Documentation Engineer for Cropchain.
Create a detailed documentation planning output that includes:
- user documentation outline
- training materials outline
- admin and API documentation outline
- release notes outline
- documentation tasks
- review tasks
- rework tasks
- productivity and effort estimates
Use 3 pages completed per day unless instructed otherwise.
""".strip(),
    llm_config=TEAM_LLM_CONFIG,
    code_execution_config=False,
    human_input_mode="NEVER",
)

print("Experiment 3 agents created successfully.")
```

```python
project_manager_to_test_engineer_prompt = f"""
The following Waterfall planning artifacts for "{PROJECT_NAME}" are complete.

=== PROJECT MANAGER SUMMARY ===
{pm_summary_text}

=== REQUIREMENTS ENGINEER OUTPUT ===
{requirements_text}

=== SYSTEM ENGINEER OUTPUT ===
{design_text}

=== SOFTWARE ENGINEER OUTPUT ===
{implementation_text}

Using these upstream artifacts, create a detailed test planning output for Cropchain.

Your output must include:
1. test plan overview
2. mapped test cases tied to important requirements
3. critical workflow coverage
4. validation for prediction, pricing, logistics visibility, alerts, and traceability
5. testing tasks and activities
6. review tasks
7. rework tasks
8. productivity estimate using 2 test cases per day
9. effort estimate in days
10. assumptions, risks, and open questions

Make the response structured and presentation-ready.
""".strip()

project_manager_to_documentation_engineer_prompt = f"""
The following Waterfall planning artifacts for "{PROJECT_NAME}" are complete.

=== PROJECT MANAGER SUMMARY ===
{pm_summary_text}

=== REQUIREMENTS ENGINEER OUTPUT ===
{requirements_text}

=== SYSTEM ENGINEER OUTPUT ===
{design_text}

=== SOFTWARE ENGINEER OUTPUT ===
{implementation_text}

Create the documentation planning output for Cropchain.

Your output must include:
1. user documentation outline
2. training materials outline
3. admin and API documentation outline
4. release notes outline
5. documentation tasks and activities
6. review tasks
7. rework tasks
8. productivity estimate using 3 pages per day
9. effort estimate in days
10. assumptions, risks, and open questions

Make the response structured and presentation-ready.
""".strip()

print("Experiment 3 prompts prepared.")
```

## Run + Save block for Experiment 3

```python
def save_text(filename: str, text: str):
    file_path = OUTPUT_DIR / filename
    file_path.write_text(text, encoding="utf-8")
    print(f"Saved: {file_path}")

def save_chat_history(filename: str, chat_result):
    file_path = OUTPUT_DIR / filename
    if hasattr(chat_result, "chat_history"):
        lines = []
        for item in chat_result.chat_history:
            role = item.get("name", item.get("role", "Unknown"))
            content = item.get("content", "")
            lines.append(f"{role}:\n{content}\n")
        file_path.write_text(("\n" + ("-" * 80) + "\n").join(lines), encoding="utf-8")
        print(f"Saved chat history: {file_path}")
    else:
        print("No chat_history found on chat result.")
```

## Run Experiment 3

This section generates:
- the Test Engineer output
- the Documentation Engineer output

After running, review the outputs before saving your final analysis notes.

```python
def run_experiment_3():
    print("=" * 80)
    print("STEP 1: Project Manager -> Test Engineer")
    print("=" * 80)

    pm_to_test = project_manager_agent.initiate_chat(
        recipient=test_engineer_agent,
        message=project_manager_to_test_engineer_prompt,
        max_turns=1,
    )

    print("\n" + "=" * 80)
    print("STEP 2: Project Manager -> Documentation Engineer")
    print("=" * 80)

    pm_to_docs = project_manager_agent.initiate_chat(
        recipient=documentation_engineer_agent,
        message=project_manager_to_documentation_engineer_prompt,
        max_turns=1,
    )

    return pm_to_test, pm_to_docs
```

```python
pm_to_test_result, pm_to_docs_result = run_experiment_3()
```

```python
save_chat_history("01_pm_to_test_engineer_chat.txt", pm_to_test_result)
save_chat_history("02_pm_to_documentation_engineer_chat.txt", pm_to_docs_result)

if hasattr(pm_to_test_result, "chat_history") and pm_to_test_result.chat_history:
    test_output = pm_to_test_result.chat_history[-1].get("content", "")
    save_text("03_test_engineer_output.md", test_output)

if hasattr(pm_to_docs_result, "chat_history") and pm_to_docs_result.chat_history:
    docs_output = pm_to_docs_result.chat_history[-1].get("content", "")
    save_text("04_documentation_engineer_output.md", docs_output)

print("Experiment 3 artifacts saved.")
```
